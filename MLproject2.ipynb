```python
plt.figure(figsize=(8, 5))
plt.hist(mall_customers['Age'], bins=10, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(mall_customers['Annual Income (k$)'], mall_customers['Spending Score (1-100)'], color='blue', alpha=0.6)
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Select only 'Annual Income' and 'Spending Score'
data = mall_customers[['Annual Income (k$)', 'Spending Score (1-100)']].corr()

# Generate heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(data, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title('Annual Income vs Spending Score Correlation')
plt.show()
```
